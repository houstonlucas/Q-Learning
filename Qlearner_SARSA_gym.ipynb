```python
import random
import numpy as np
import gym
from numpy.random import choice
```

```python
def normalizedList(size):
    l = [random.random() for i in range(size)]
    s = sum(l)
    lnorm = [x/s for x in l]
    return lnorm
```

```python
class Qlearner():
    def __init__(self, env, gamma = 0.9, alpha = 0.1):
        self.q = np.random.rand(env.observation_space.n, env.action_space.n)
        self.alpha = alpha
        self.env = env #mdp
        self.env.reset()
        self.score = 0
        self.prevAction = 0
        self.state = 0 
        self.gamma = gamma
        
    def exploit(self, epsilon = 0.1):
        """"Figure out what to do based upon value function"""
        random = np.random.random()
        prevState = self.state
        #Epsilon greedy
        if(random < epsilon):
            #Behave randomly
            actionTaken = self.env.action_space.sample() # Picks a random value from array
            
        else:
            actionTaken = np.argmax(self.q[self.state])
            # print(actionTaken)
        sNew, reward, done, info = self.env.step(actionTaken) #returns s, r, d?, info
        self.score += reward - 0.01
        self.q[self.state][actionTaken] += self.alpha*(reward + self.gamma*(self.q[self.state][actionTaken]) \
                                                      - self.q[prevState][self.prevAction])
        self.prevAction = actionTaken
        self.state = sNew #we somehow want to update the state to be the one we've just moved into
        
        if done and reward == 0:
            reward = -1
        
        return reward, done
```

```python
envname = 'FrozenLakeMod-v0'
env = gym.make(envname)

maxItters = 1000

qlearner = Qlearner(env)
trainItters = 10000

for i in range(trainItters):
    t = 0
    done = False
    env.reset()
    while not done:
        t+=1
        rew, done = qlearner.exploit(epsilon=1.0)
        if t > maxItters:
            done = True
    
print(qlearner.score / trainItters)

qlearner.score = 0

testItters = 1
for i in range(testItters):
    t = 0
    done = False
    env.reset()
    while not done:
        t+=1
        rew, done = qlearner.exploit(epsilon=0.01)
        env.render()
        if t > maxItters:
            done = True
    
print(qlearner.score/testItters)

# rew = qlearner.exploit()
# print(rew)
```
